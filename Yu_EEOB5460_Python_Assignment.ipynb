Valerie (me): Explanation for each code line is commented below the code. In sum, this function reads a FASTA file and returns a dictionary of the sequences indexed by species name. The input is `fasta_fn` and the return is a dictionary. In the dictionary, each species name has a value (corresponding sequence).

```python
def get_sequences_from_file(fasta_fn): 
    # This function reads a FASTA file, `fasta_fn`.
    sequence_data_dict = {} 
    # Here, it is creating an empty dictionary. 
    for record in SeqIO.parse(fasta_fn, "fasta"): 
    # For every `record`, aka sequence entry, SeqIO (Biopython Library) reads through each entry in the `fasta_fn` FASTA file.
        description = record.description.split()
        # Splits the text string description in the record into a list of smaller substrings.
        species_name = description[1] + " " + description[2]
        # Makes `species_name` the second and third word (since we have zero-based index in Python). 
        # There will be a space between I'm assuming the genus and species
        sequence_data_dict[species_name] = record.seq
        # `species_name` is stored into the dictionary Dr. X created, with `record.seq` being the sequence values.
    return(sequence_data_dict)
    # Returns the now filled dictionary.
```

## 2. YOUR STRING-TRANSLATE FUNCTION
Write a function that translates sequences. All sequences start at codon position 1. Complete a function that translates using a loop over the string of nucleotides. Here is  some pseudo-code and suggestions. Feel free to change the function and variable names

Valerie: Like above, my explanation is below the code. I replaced the pseudo-code.

This function takes the sequence `string_nucleotides` and reads each group of three nucleotides (codons). It then translates each codon into amino acids using the Vertebrate Mitochondrial Codon Table. Any incomplete codons at the end are skipped and the loop is stopped if it runs into a stop codon. Returns an amino acid sequence as a string.

```python
def translate_function(string_nucleotides):
    mito_table = CodonTable.unambiguous_dna_by_name["Vertebrate Mitochondrial"]
    # This should work using BioPython and I will check what this returns to make sure.
    
    aa_seq_string = ""
    # I am making a for-loop. Following Dr. X's example, I need to create an empty string for my amino acids.

    for i in range(0, len(string_nucleotides), 3):
    # This is the for-loop. I use `range(start at 0, stop at end aka the length of sequence, step is 3):`
    # Codons are made of 3 nucleotides.  
  
        codon = str(string_nucleotides[i:i+3])
        # Unlike in R, Python has zero-based indexing and cuts before end, so I use i+3 to include all 3 nucleotides.

        if len(codon) < 3:
            continue
        # This is saying "if the length codon is less than 3, aka not a full codon at the end, skip it."
        
        if codon in mito_table.stop_codons:
            break
        # This is saying "if the codon is one of the stop codons, break", with break = exit loop
        
        amino_acid = mito_table.forward_table[codon]
        # mito_table.forward_table[codon] allows us to retrieve the amino acid
        aa_seq_string = aa_seq_string + amino_acid
        # then I add the amino acid onto the protein sequence `aa_seq_string`
        
    return aa_seq_string

# mito_table = CodonTable.unambiguous_dna_by_name["Vertebrate Mitochondrial"]
# print(mito_table)
# print(mito_table.stop_codons)
# print(mito_table.forward_table)
```

## 3. YOUR ALTERNATIVE FUNCTION

Valerie: `Getting Started Tips` says use .translate() for amino acid sequences. That's in the BioPython library. `.translate()` will "turn a nucleotide sequence into a protein sequence by creating a new Seq object". This means no for-loop.

```python
# I need to tell it which table to reference. It also has a `to_stop` argument for stop codons.
def biopython_translate_function(string_nucleotides): 
    biopython_aa_sequence = string_nucleotides.translate(table = "Vertebrate Mitochondrial", to_stop = True)
    # I set it to true since I don't want the stop symbols in my string.
    return str(biopython_aa_sequence)
```

## 5.  YOUR GC CONTENT ANALYSIS FUNCTION
Write a function that calculates the GC-content (proportion of "G" and "C") of each DNA sequence and returns this value.

```python
def compute_GC_content(DNA_seq):
    DNA_seq = str(DNA_seq)
    # Just incase it's not a string.
    G_count = DNA_seq.count("G")
    C_count = DNA_seq.count("C")
    # These give me separate variable for G and C.
    GC_content = (G_count + C_count) / len(DNA_seq)
    # We want proportion of G and C to total amount of nucleotides.
    return GC_content
```

## 7. 
### Write a for-loop that translates each sequence and also gets molecular weight and computes the GC content of each translated sequence and adds those data to DataFrame 

```python
for species_name, DNA_seq in cytb_seqs.items():
    aa_seq = biopython_translate_function(DNA_seq)
    # Using alternative function 3 to translate sequence.
    molecular_weight = compute_molecular_weight(aa_seq)
    # Using function 4 to calculate molecular weight of each amino acid sequence.
    GC_content = compute_GC_content(DNA_seq)
    # Using function 5 to find GC content.

    penguins_df.loc[penguins_df["species"] == species_name, "molecular weight"] = molecular_weight
    penguins_df.loc[penguins_df["species"] == species_name, "GC content"] = GC_content
    # To fill in empty cells in DF, I use `.loc` to select column names.
    
penguins_df.head()


```
